```python
import pandas as pd
```

```python

def cargar_datos(sheet_name):
      return pd.read_excel("Matriz_de_adyacencia_data_team.xlsx", sheet_name=sheet_name, index_col=False)
```

```python

df = cargar_datos("Matriz de adyacencia")

df.head()
```

```python
# iterate through the superior triangle of the matrix
for i in range(0, len(df)):
    for j in range(i+1, len(df)):
        if df.iloc[i, j] == 1:
          print("INSERT INTO matriz_adyacencia (id, id_adyacente) VALUES")
          print(f"({i+1}, {j}),")
```
